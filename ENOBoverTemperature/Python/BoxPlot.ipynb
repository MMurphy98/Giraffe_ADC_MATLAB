# Box Plot for Calibration Statistics


## Loading Data from .mat file

```python
import scipy.io as scio
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
```

```python
data=scio.loadmat('../DATA_PROCESS_Calibration_Results_Data.mat')

type(data)
data.keys()
```

```python
box_data = np.array(data['data'])
```

```python
plt.boxplot?
```

```python
box_label = ['x1', 'x5', 'x10', 'x50', 'x100']
plt.grid(True)
plt.boxplot(
    box_data,
    medianprops={'color': 'red', 'linewidth': '1.5'},
    meanline=True,
    showmeans=True,
    meanprops={'color': 'blue', 'ls': '--', 'linewidth': '1.5'},
    flierprops={"marker": "", "markerfacecolor": "red", "markersize": 10},
    labels=box_label
)
plt.show()
```
